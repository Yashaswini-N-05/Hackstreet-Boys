```python
import math
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.layers import Dense, LSTM, Dropout
from keras.models import Sequential
from sklearn.metrics import accuracy_score

plt.style.use('fivethirtyeight')

#Downloading the stock data
df = yf.download('AMZN', start = '2010-01-01', end = "2024-10-18")

#Plotting the close price of AMZN
plt.figure(figsize = (12,8))
plt.title('AMZN Stock price')
plt.plot(df['Close'])
plt.xlabel("Date")
plt.ylabel("USD")
plt.show()

data = df.filter(['Close'])
dataset = data.values
training_data_len = math.ceil(len(dataset) * 0.8)
len(dataset)

#Scaling the data
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)
train_data = scaled_data[:training_data_len]
len(train_data)

#Creating training data
x_train = []
y_train = []
timestep = 60
for i in range(timestep, len(train_data)):
  x_train.append(train_data[i-60:i, 0])
  y_train.append(train_data[i,0])

len(x_train)

#Reshaping training data to fit into LSTM
x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
x_train.shape

#Creating test data
test_data = scaled_data[training_data_len-60:,:]
x_test = []
y_test = dataset[training_data_len : , :]
for i in range(timestep, len(test_data)):
  x_test.append(test_data[i-timestep:i, 0])

#Reshaping the test data after converting to a numpy array
x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

#Creating an LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape = (x_train.shape[1], x_train.shape[2] )))
model.add(LSTM(50, return_sequences= False))
model.add(Dense(25))
model.add(Dense(1))

model.compile(optimizer ='adam', loss = 'mean_squared_error')

#Training the LSTM model
model.fit(x_train, y_train, epochs = 15, batch_size=32)

#Model prediction
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

#Accuracy measurement
rmse = np.sqrt(np.mean(predictions - y_test)**2)
rmse

#Creating the plotting data
train = data[:training_data_len]
valid = data[training_data_len:]
valid['Predictions'] = predictions

#Future prediction
future_days = 365
last_sequence = scaled_data[-60:]  # Get the last 60 days of scaled data
future_predictions = []

for _ in range(future_days):
    last_sequence_reshaped = np.reshape(last_sequence, (1, last_sequence.shape[0], 1))
    predicted_price = model.predict(last_sequence_reshaped)
    future_predictions.append(predicted_price[0, 0])
    last_sequence = np.append(last_sequence, predicted_price, axis=0)[-60:]

# Inverse transform future predictions
future_predictions_inv = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

# Creating DataFrame for future predictions
future_dates = pd.date_range(start=df.index[-1] + pd.Timedelta(days=1), periods=future_days)
future_prices = pd.DataFrame(future_predictions_inv, index=future_dates, columns=['Future Predictions'])

#Creating the graph
plt.figure(figsize = (16,8))
plt.title('Actual v/s Predicted price')
plt.xlabel('Date')
plt.ylabel('USD')
plt.plot(train['Close'], linewidth=1.5)
plt.plot(valid[['Close', 'Predictions']], linewidth=2)
plt.plot(future_prices, linewidth=1.8, linestyle='--', label='Future Predictions')
plt.legend(['Train','Val','Predictions'], loc='upper left')
plt.show()
```
